```python
import torch

%load_ext rich
```

```python
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad
```

```python
y = 2*torch.dot(x,x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4*x
```

```python
x.grad.zero_()

y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x*x
y.backward(gradient=torch.ones(len(y)))
x.grad
```
